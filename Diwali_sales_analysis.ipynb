# **Exploratory Data Analysis📊**

```python
df.columns
```

```python
plot = sns.countplot(x='Gender', data=df, hue='Gender')
plt.title = plt.gca().set_title('Total Buyers')

for bars in plot.containers:
  plot.bar_label(bars)
```

```python
sales = df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
plot = sns.barplot(x = 'Gender', y = 'Amount', data = sales, hue='Gender')
```

## From above graphs we can conclude
*   most of buyers are Females
*   purchasing power of Females are greater than men

## **AGE**

```python
plot = sns.countplot(data = df, x = 'Age Group')

for bars in plot.containers:
  plot.bar_label(bars)
```

```python
plot = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in plot.containers:
  plot.bar_label(bars)
```

```python
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plot = sns.barplot(x='Age Group', y='Amount', hue = 'Age Group', data=sales_age)
```

### Most Buyers are of age group between ***26-35 yrs Female***

## **STATE**

```python
# total number of orders from top 5 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(5)
sns.set(rc={'figure.figsize':(8,4)})
# create bar plot
plot = sns.barplot(data=sales_state, x='State', y='Orders', hue='Orders')
plt.title = plt.gca().set_title('Top 5 States')

# add labels to bars
for bars in plot.containers:
  plot.bar_label(bars)
```

```python
# total number of orders from bottom 5 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).tail(5)
sns.set(rc={'figure.figsize':(8,4)})
plot = sns.barplot(data=sales_state, x='State', y='Orders', hue='State', palette='Reds')
plt.title = plt.gca().set_title('Bottom 5 States')

for bars in plot.containers:
  plot.bar_label(bars)
```

### *Most of orders are from **UttarPradesh, Maharashtra and Karnataka** respectively.*

## **MARITAL STATUS**

```python
plot = sns.countplot(data=df, x='Marital_Status', hue='Marital_Status', legend=False)
for bars in plot.containers:
  plot.bar_label(bars)

sns.set(rc={'figure.figsize':(4,4)})
```

```python
sales_state= df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
plot = sns.barplot(data=sales_state, x='Marital_Status', y='Amount', hue='Gender')
```

### *Most buyers are **Married(Women)** and have high purchasing power*

## **OCCUPATION**

```python
# people count based on occupation
sns.set(rc={'figure.figsize':(22,5)})
plot = sns.countplot(data=df, x='Occupation', hue='Occupation')

for bars in plot.containers:
  plot.bar_label(bars)
```

```python
sales_occupation = df.groupby(['Occupation'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=True)

sns.set(rc={'figure.figsize':(22,5)})
plot = sns.barplot(data = sales_occupation, x = 'Occupation', y = 'Orders', hue='Occupation', palette = 'Blues')
```

### *Most buyers are working in **IT, HealthCare and Aviation** sector*

## **CATEGORY**

```python
#Product Category generate revenue
sales_product = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

plt.figure(figsize=(20,5))
sns.set(style='darkgrid')
plt.pie(sales_product['Amount'], labels = sales_product['Product_Category'], autopct='%0.2f%%')
plt.show()
```

```python
# the top 5 states and top 3 categories by order count
top_5_age = df.groupby('Age Group')['Orders'].sum().nlargest(6).index
top_3_category = df.groupby('Product_Category')['Orders'].sum().nlargest(3).index
# Filter data for top 5 states and top 3 categories
filtered_df = df[df['Age Group'].isin(top_5_age) & df['Product_Category'].isin(top_3_category)]
# Group by State and Product_Category and sum Amount
sales_state_category = filtered_df.groupby(['Age Group', 'Product_Category'] , as_index=False)['Orders'].sum()
plt.figure(figsize=(10, 4))
plot = sns.barplot(x='Age Group', y='Orders', hue='Product_Category', data=sales_state_category)
for bars in plot.containers:
  plot.bar_label(bars)
```

```python
# the top 5 states and top 3 categories by order count
top_5_states = df.groupby('State')['Amount'].sum().nlargest(5).index
top_3_category = df.groupby('Product_Category')['Amount'].sum().nlargest(3).index

# Filter data for top 5 states and top 3 categories
filtered_df = df[df['State'].isin(top_5_states) & df['Product_Category'].isin(top_3_category)]

# Group by State and Product_Category and sum Amount
sales_state_category = filtered_df.groupby(['State', 'Product_Category'] , as_index=False)['Amount'].sum()

plt.figure(figsize=(10, 4))
plot = sns.lineplot(x='State', y='Amount', hue='Product_Category', data=sales_state_category)
```

```python
# top 10 most sold products
fig1, plot = plt.subplots(figsize=(12,5))
df.groupby('Product_ID')['Orders'].sum().nlargest(10).plot(kind='bar')
```
